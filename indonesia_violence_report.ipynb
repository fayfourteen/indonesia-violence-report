```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as pyo
pyo.init_notebook_mode(connected=True)
import warnings
warnings.filterwarnings('ignore')
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('indonesia-violence-reporting-text.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# Countplot for Layanan Hukum (Legal Services)
LayananHukum_colors = ['chartreuse', 'magenta']
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Layanan Hukum', palette=LayananHukum_colors)
plt.title('Layanan Hukum (Legal Services)')
plt.show()
```

```python
# Countplot for Layanan Medis (Medical Services)
LayananMedis_colors = ['coral', 'dodgerblue']
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Layanan Medis', palette=LayananMedis_colors)
plt.title('Layanan Medis (Medical Services)')
plt.show()
```

```python
layanan_psikologis_counts = df['Layanan Psikologis'].value_counts()
labels = layanan_psikologis_counts.index
values = layanan_psikologis_counts.values

plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Layanan Psikologis (Psychological Services)')
plt.axis('equal')
plt.show()
```

```python
# Donut chart for Layanan Psikologis (Psychological Services)
LayanaPsikologis_df = df['Layanan Psikologis'].value_counts()
labels_psikologis = LayanaPsikologis_df.index
values_psikologis = LayanaPsikologis_df.values

fig_psikologis = go.Figure(data=[go.Pie(labels=labels_psikologis, values=values_psikologis, hole=.6)])
fig_psikologis.update_layout(title_text='Layanan Psikologis (Psychological Services)')
fig_psikologis.show()
```

```python
# Pie chart for Jaminan Keselamatan (Safety Guarantee)
jaminan_keselamatan_counts = df['Jaminan Keselamatan'].value_counts()
labels_keselamatan = jaminan_keselamatan_counts.index
values_keselamatan = jaminan_keselamatan_counts.values
plt.figure(figsize=(8, 8))
plt.pie(values_keselamatan, labels=labels_keselamatan, autopct='%1.1f%%', startangle=140)
plt.title('Jaminan Keselamatan (Safety Guarantee)')
plt.axis('equal')
plt.show()
```

```python
# Pie chart for Layanan Pendidikan (Education Services)
layanan_pendidikan_counts = df['Layanan Pendidikan'].value_counts()
labels_pendidikan = layanan_pendidikan_counts.index
values_pendidikan = layanan_pendidikan_counts.values
plt.figure(figsize=(8, 8))
plt.pie(values_pendidikan, labels=labels_pendidikan, autopct='%1.1f%%', startangle=140)
plt.title('Layanan Pendidikan - Education Services')
plt.axis('equal')
plt.show()
```

```python
# Pie chart for Pengasuhan Pengganti (Substitute Care) - PENGGANTI DONUT
pengasuhan_pengganti_counts = df['Pengasuhan Pengganti'].value_counts()
labels_pengganti = pengasuhan_pengganti_counts.index
values_pengganti = pengasuhan_pengganti_counts.values
plt.figure(figsize=(8, 8))
plt.pie(values_pengganti, labels=labels_pengganti, autopct='%1.1f%%', startangle=140)
plt.title('Pengasuhan Pengganti (Substitute Care)')
plt.axis('equal')
plt.show()
```

```python
# Countplot for Bantuan Sosial (Social Assistance)
BantuanSosial_colors = ['coral', 'dodgerblue']
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Bantuan Sosial', palette=BantuanSosial_colors)
plt.title('Bantuan Sosial (Social Assistance)')
plt.show()
```

```python
# Word Cloud for Text
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color='black',
                      height=2000,
                      width=2000,
                      colormap="Set2",
                      stopwords=stopwords).generate(str(df["Text"]))
plt.rcParams['figure.figsize'] = (12, 12)
plt.axis("off")
plt.imshow(wordcloud)
plt.title("Indonesia Violence Reporting Text")
plt.show()
```
